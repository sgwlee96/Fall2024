```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import LabelEncoder
```

## Wave 3 (1988)

```python
df = pd.read_csv(f"{base_path}/df_1988.csv", low_memory=False)
df.head()
```

```python
for col in df.columns:
    print(col)
```

```python
target_column = "marital_instability_in_1988"
year_of_wave = "1988"
previous_wave_years = ["1980", "1983"]
fig_path = f"{root_path}/{year_of_wave}_figs"
os.makedirs(fig_path, exist_ok=True)
```

```python
# since this is the third wave, we only look at respondants from this wave that have not dropped off since the last one
df.duplicated().sum()
```

```python
df.drop_duplicates(inplace=True)
```

```python
# we dont need the name of the respondants for analysis
col_selector = [col for col in df.columns if "name" in col]
df.drop(columns=col_selector, inplace=True)
df.head()
```

```python
# which columns relate to other people besides respondant
other_person_cols = [col for col in df.columns if "person" in col]
other_person_cols = [
    col for col in other_person_cols if any(str(i) in col for i in range(11))
]
other_person_cols
```

```python
# if a majority of respondants dont have x number of other people, we dont need that for analysis
for col in [col for col in other_person_cols if "relation" in col]:
    print(df[col].value_counts() / len(df) * 100)
    print()
```

```python
# Person 5 and onwards are 80% non existant
col_selector = [
    col for col in other_person_cols if any(str(i) in col for i in range(5, 11))
]
df.drop(columns=col_selector, inplace=True)
```

```python
# dropping columns relating to previous studies
previous_waves = [
    col for col in df.columns if any(i in col for i in previous_wave_years)
]
df.drop(columns=previous_waves, inplace=True)
changes = [col for col in df.columns if "change" in col]
changes = changes[:-3]
df.drop(columns=changes, inplace=True)
```

```python
df.info()
```

```python
# convert to numerical type
for col in df.columns:
    df[col] = df[col].astype(np.float32, errors="ignore")
```

```python
# still many object types
df.info()
```

```python
# According to the codebook, many values can be considered nullable
# such as "DK", "INAP", "REF", "97", "997", etc
possible_nan = [
    "DK",
    "INAP",
    "INAPP",
    "REF",
    "9",
    "97",
    "997",
    "9997",
    "99997",
    "999997",
]
df.replace(
    to_replace=possible_nan,
    value=None,
    inplace=True,
)
```

```python
# try converting to numerical
for col in df.columns:
    df[col] = df[col].astype(np.float32, errors="ignore")
```

```python
# better
df.info()
```

```python
for k, v in pd.Series(df.values.flatten()).value_counts().items():
    print(f"{k:<50} |  {v}")
```

```python
# which columns could contain nullable values
possible_nan.extend([97.0, 997.0, 9997.0])
check_nan_cols = []
for col in df.columns:
    is_nan = set(df[col].values).intersection(set(possible_nan))
    if is_nan:
        print(f"{col:<50} | {is_nan}")
        check_nan_cols.append(col)
```

```python
# by inspection and refering to code book, we can see all the above columns are using 997, 9997, etc as nulls
# replacing them
for col in df.columns:
    is_nan = set(df[col].values).intersection(set(possible_nan))
    if is_nan:
        df[col] = df[col].replace(to_replace=is_nan, value=None)
```

```python
# converting any remaning cols to numeric
for col in df.columns:
    df[col] = df[col].astype(np.float32, errors="ignore")
```

```python
df.info()
```

```python
# checking for any other values to change
for k, v in pd.Series(df.values.flatten()).value_counts().items():
    print(f"{k:<50} |  {v}")
```

```python
# many columns are actually the same question
col_selector = [col for col in df.columns if ":" in col]
col_selector
```

```python
col_selector = [col for col in df.columns if "divorce_worries:" in col]
print(col_selector)
print()
pd.Series(df[col_selector].values.flatten()).value_counts()
```

```python
title = f"Divorce worries ({year_of_wave})"
pd.DataFrame(
    {
        col.split(":")[1].replace("_", " ").strip(): df[col].value_counts()
        for col in col_selector
    }
).fillna(0).plot(
    kind="barh",
    figsize=(10, 6),
    title=title,
    xlabel="Count",
    ylabel="Answer",
    grid=True,
    width=0.8,
)
plt.tight_layout()
plt.savefig(f"{fig_path}/{title}.png")
plt.show()
```

```python
le = LabelEncoder()
tmp_df = df[[*col_selector, target_column]].copy()
for col in col_selector:
    tmp_df[col] = le.fit_transform(tmp_df[col])
tmp_df[target_column].fillna(0, inplace=True)
tmp_df.rename(
    columns={col: col.split(":")[1].replace("_", " ").strip() for col in col_selector},
    inplace=True,
)
correlations = tmp_df.corr()[target_column].drop(target_column)
plt.figure(figsize=(10, 6))
sns.barplot(x=correlations.index, y=correlations.values, palette="viridis")
title = f"Correlation of divorce worries with {target_column} ({year_of_wave})"
plt.title(title)
plt.xlabel("Columns")
plt.ylabel("Correlation Coefficient")
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig(f"{fig_path}/{title}.png")
plt.show()
```
