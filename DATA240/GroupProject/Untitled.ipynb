```python
related_columns = [col for col in df97.columns if 'divorce' in col.lower() or 'marital' in col.lower()]
print(related_columns)
```

```python
corr_df97 = df97[related_columns]
```

```python
from sklearn.preprocessing import LabelEncoder

pd.set_option('display.max_rows', None)
# pd.reset_option('display.max_rows')


label_encoder = LabelEncoder()
for column in corr_df97.select_dtypes(include=['object']).columns:
    corr_df97[column] = label_encoder.fit_transform(corr_df97[column].astype(str))

# Calculate correlation with marital_instability_1997
correlations = corr_df97.corr()['marital_instability_1997'].drop('marital_instability_1997')
result_corr_1997 = pd.DataFrame(correlations)
```

```python
result_corr_1997.sort_values(by='marital_instability_1997', ascending=False)
```

```python
# Sort by absolute values of 'marital_instability_1997'
sorted_corr = result_corr_1997.reindex(result_corr_1997['marital_instability_1997'].abs().sort_values(ascending=False).index)

# Reset index (optional)
# sorted_corr.reset_index(drop=True, inplace=True)

pd.DataFrame(sorted_corr)
```
