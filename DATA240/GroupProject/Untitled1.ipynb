## Trend family Income - Marital Instability

```python
year_to_marital_col = {
    1980: 'marital_instability_1980_(1980)',
    1983: 'marital_instability_1983_(1983)',
    1988: 'marital_instability_in_1988_(1988)',
    1992: 'marital_stability_1992_(1992)', 
    1997: 'marital_instability_1997_(1997)',
    2000: 'marital_instability_2000_(2000)'
}
```

```python
# Dropping rows with NaNs in relevant columns
for col in family_income_20000_cols:
    year = int(col.split('(')[-1][:-1])  # Extract the year from the column name
    marital_instability_col = year_to_marital_col.get(year)  # Get the corresponding marital instability column
    
    # Only drop rows if the marital instability column exists for this year
    if marital_instability_col:
        df.dropna(subset=[col, marital_instability_col], inplace=True)
```

```python
sorted_columns = sorted(family_income_20000_cols, key=lambda col: int(col.split('(')[-1][:-1]))

print(sorted_columns)
```

```python
results = {}
for col in sorted_columns:
    year = int(col.split('(')[-1][:-1])  # Extract year as an integer
    marital_instability_col = year_to_marital_col.get(year)  # Get marital instability column
    if marital_instability_col:
        # Standardize income labels
        df[col] = df[col].str.replace(',', '').str.replace('> $20000', '$20000 OR MORE').str.replace('< $20000', 'LESS THAN $20000')
        
        # Group by income level and calculate the mean marital instability for the year
        grouped = df.groupby(col)[marital_instability_col].mean()
        results[year] = grouped
```

```python
# Convert results into a DataFrame
trend_df = pd.DataFrame(results).T  # Transpose to have years as rows
trend_df.index.name = 'Year'
trend_df.reset_index(inplace=True)

# Rename income columns for consistency
trend_df = trend_df.rename(columns={
    '$20,000 OR MORE': '$20000 OR MORE',
    'LESS THAN $20,000': 'LESS THAN $20000'
})

# Combine duplicate columns if any exist
trend_df = trend_df.groupby(axis=1, level=0).sum()
trend_df['Year'] = trend_df['Year'].astype(int).astype(str)
trend_df['Year']
```

```python
import matplotlib.pyplot as plt

trend_df.plot(x='Year', kind='line', marker='o')
plt.title("Marital Instability Trends by Income Level")
plt.ylabel("Average Marital Instability")
plt.xlabel("Year")
plt.legend(title="Income Levels")
plt.grid()
plt.tight_layout()
plt.show()
```

## Income category (more detail) and marital instability Trend

### Year 2000

```python
income_category_2000 = ['income_category_-_more_than_$20,001_(2000)',
 'income_category_less_than_$20,001_(2000)']

for i in income_category_2000:
    print()
    print(df[income_category_2000][i].value_counts())
```

```python
income_mapping = {
    '$100,000 OR MORE': 100000,
    '$90,000 - $99,999': 95000,
    '$80,000 - $89,999': 85000,
    '$70,000 - $79,999': 75000,
    '$60,000 - $69,999': 65000,
    '$50,000 - $59,999': 55000,
    '$40,000 - $49,999': 45000,
    '$30,000 - $39,999': 35000,
    '$25,000 - $29,999': 27500,
    '$20,000 - $24,999': 22500,
    '$15,000 - $19,999': 17500,
    '$10,000 - $14,999': 12500,
    'UNDER $5,000': 2500
}

df['income_more_than_20k_2000'] = df['income_category_-_more_than_$20,001_(2000)'].map(income_mapping)
df['income_less_than_20k_2000'] = df['income_category_less_than_$20,001_(2000)'].map(income_mapping)
```

```python
# df.drop(columns = ['income_more_than_20k','income_less_than_20k'], inplace=True)
df['income_combined'] = df[['income_more_than_20k_2000', 'income_less_than_20k_2000']].sum(axis=1, skipna=True)
# df['income_combined'].value_counts()

income_instability = df.groupby('income_combined')['marital_instability_2000_(2000)'].mean().reset_index()
```

```python
reverse_income_mapping = {v:k for k, v in income_mapping.items()}
income_instability['income_label'] = income_instability['income_combined'].map(reverse_income_mapping)
income_instability
```

```python
# Drop rows with NaN income labels, as they can't be plotted meaningfully
income_instability_cleaned = income_instability.dropna(subset=['income_label'])

# Plot the cleaned data
plt.figure(figsize=(12, 6))
plt.plot(
    income_instability_cleaned['income_label'],
    income_instability_cleaned['marital_instability_2000_(2000)'],
    marker='o',
    linestyle='-',
    color='b'
)

# Add labels, title, and grid
plt.xlabel('Income Category')
plt.ylabel('Mean Marital Instability')
plt.title('Trend Between Income and Marital Instability (2000)')
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.grid()

plt.show()
```
