## Trend family Income - Marital Instability

```python
year_to_marital_col = {
    1980: 'marital_instability_1980_(1980)',
    1983: 'marital_instability_1983_(1983)',
    1988: 'marital_instability_in_1988_(1988)',
    1992: 'marital_stability_1992_(1992)', 
    1997: 'marital_instability_1997_(1997)',
    2000: 'marital_instability_2000_(2000)'
}
```

```python
# Dropping rows with NaNs in relevant columns
for col in family_income_20000_cols:
    year = int(col.split('(')[-1][:-1])  # Extract the year from the column name
    marital_instability_col = year_to_marital_col.get(year)  # Get the corresponding marital instability column
    
    # Only drop rows if the marital instability column exists for this year
    if marital_instability_col:
        df.dropna(subset=[col, marital_instability_col], inplace=True)
```

```python
sorted_columns = sorted(family_income_20000_cols, key=lambda col: int(col.split('(')[-1][:-1]))

print(sorted_columns)
```

```python
results = {}
for col in sorted_columns:
    year = int(col.split('(')[-1][:-1])  # Extract year as an integer
    marital_instability_col = year_to_marital_col.get(year)  # Get marital instability column
    if marital_instability_col:
        # Standardize income labels
        df[col] = df[col].str.replace(',', '').str.replace('> $20000', '$20000 OR MORE').str.replace('< $20000', 'LESS THAN $20000')
        
        # Group by income level and calculate the mean marital instability for the year
        grouped = df.groupby(col)[marital_instability_col].mean()
        results[year] = grouped
```

```python
# Convert results into a DataFrame
trend_df = pd.DataFrame(results).T  # Transpose to have years as rows
trend_df.index.name = 'Year'
trend_df.reset_index(inplace=True)

# Rename income columns for consistency
trend_df = trend_df.rename(columns={
    '$20,000 OR MORE': '$20000 OR MORE',
    'LESS THAN $20,000': 'LESS THAN $20000'
})

# Combine duplicate columns if any exist
trend_df = trend_df.groupby(axis=1, level=0).sum()
trend_df['Year'] = trend_df['Year'].astype(int).astype(str)
trend_df['Year']
```

```python
import matplotlib.pyplot as plt

trend_df.plot(x='Year', kind='line', marker='o')
plt.title("Marital Instability Trends by Income Level")
plt.ylabel("Average Marital Instability")
plt.xlabel("Year")
plt.legend(title="Income Levels")
plt.grid()
plt.tight_layout()
plt.show()
```

## Marriage Problem

```python
df_marriage = df_income.copy()
```

```python
all_cols = sorted(df_marriage.columns)
orig_marriage_problem_cols = [
    *[col for col in all_cols if "marriage_problem:" in col],
    *[col for col in all_cols if "marriage_problem-" in col],
    *[col for col in all_cols if "marriage_problem_:" in col],
    *[col for col in all_cols if "marriage_problem_-" in col],
]
marriage_problem_cols = orig_marriage_problem_cols.copy()
marriage_problem_cols = [x[len("marriage_problem:"):] for x in marriage_problem_cols]
marriage_problem_cols = [x.replace("_", " ").strip() for x in marriage_problem_cols]
marriage_problem_cols = [x.replace("w/", "with").strip() for x in marriage_problem_cols]
marriage_problem_cols[marriage_problem_cols.index("sex relations with othe (1980)")] = "sex with others (1980)"
marriage_problem_cols[marriage_problem_cols.index("is jealous (1980)")] = "jealousy (1980)"
marriage_problem_cols[marriage_problem_cols.index("one not talking to othe (1980)")] = "not talking (1980)"
marriage_problem_cols[marriage_problem_cols.index("one not at home enough (1980)")] = "not home enough (1980)"
marriage_problem_cols[marriage_problem_cols.index("gets angry easily (1983)")] = "get angry easily (1983)"
marriage_problem_dict = {}
marriage_problem_keep = [
    "get angry easily",
    "sex with others",
    "jealousy",
    "not talking",
    "not home enough"
]
```

```python
for i, x in enumerate(marriage_problem_cols):
    k = x.split("(")[0].strip()
    if k not in marriage_problem_keep:
        continue
    if k not in marriage_problem_dict.keys():
        marriage_problem_dict[k] = [(x.split("(")[1].split(")")[0], i)]
    else:
        marriage_problem_dict[k].append((x.split("(")[1].split(")")[0],i))
marriage_problem_dict
marriage_problem_col_renamer = {}
for i, k in enumerate(marriage_problem_cols):
    kc = k.split("(")[0].strip()
    if kc not in marriage_problem_keep:
        continue
    for y, j in marriage_problem_dict[kc]:
        marriage_problem_col_renamer[orig_marriage_problem_cols[j]] = f"{kc} ({y})"
marriage_problem_col_renamer
df_marriage.rename(columns=marriage_problem_col_renamer, inplace=True)
```

```python
import numpy as np

# Function to convert categorical responses to numeric
def preprocess_column(column):
    return column.replace({'YES': 1, 'NO': 0, 'YES, SELF' : 1, 'YES,SELF': 1, 'YES SELF': 1, 'YES,SPOUSE': 1, 'YES, SPOUSE' : 1, 'YES SPOUSE' : 1, 'NO,SELF': 0, 'NO SPOUSE': 0, 'BOTH' : 2 }).astype(float)

# Define years and relevant column mappings
years = ['1983', '1988', '1992', '2000']
angry_cols = [f'get angry easily ({year})' for year in years]
instability_cols = [f'marital instability ({year})' for year in years]
income_col = 'income_combined_{year}'

# Prepare a dataframe to store correlations
correlation_data = []

for year, angry_col, instability_col in zip(years, angry_cols, instability_cols):
    yearly_data = df[[angry_col, instability_col, f'income_combined_{year}']].copy()
    yearly_data.columns = ['angry', 'instability', 'income']
    
    # Preprocess columns to numeric
    yearly_data['angry'] = preprocess_column(yearly_data['angry'])
    yearly_data['instability'] = pd.to_numeric(yearly_data['instability'], errors='coerce')
    
    # Drop rows with NaN values
    yearly_data = yearly_data.dropna()
    
    # Calculate correlations for each income group
    grouped = yearly_data.groupby('income').apply(
        lambda group: group['angry'].corr(group['instability'])
    ).reset_index()
    
    grouped.columns = ['income', 'correlation']
    grouped['year'] = year
    correlation_data.append(grouped)

# Combine all yearly data
correlation_trends = pd.concat(correlation_data)

# Visualization
plt.figure(figsize=(12, 6))
sns.lineplot(data=correlation_trends, x='year', y='correlation', hue='income', marker='o')
plt.title('Correlation between "Get Angry Easily" and Marital Instability Across Years')
plt.xlabel('Year')
plt.ylabel('Correlation')
plt.legend(title='Income Range', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True)
plt.tight_layout()
plt.show()
```
