## 2. Transformation

#### Converting Mileage

Since the dominant unit in mileage column is kmpl, we will convert other units to kmpl.

```python
df['Mileage_unit'] = df['Mileage'].str.extract(r'([a-zA-Z/ ]+)$')
```

```python
df["Mileage_Num"] = df['Mileage'].str.extract(r'([0-9.]+)')
df['Mileage_Num'] = pd.to_numeric(df['Mileage_Num'], errors='coerce')
```

Now we have converted Mileage_Num column into numerical type, we should also care of the fuel type if it is CNG or LPG. 1 kg of CNG can be converted to 1.689 Liters and 1 kg of LPG can be converted to 1.96 Liters. Thus, we multiply 1.689 to the Mileage_Num of which fuel type is CNG and multiply 1.96 to which fuel type is LPG.

References:

CNG : https://www.gov.uk/hmrc-internal-manuals/gas-for-road-fuel-use/hcogas400350#:~:text=%2D%20Liquefied%20Natural%20Gas%20%2D%201%20litre,1%20Kilogram%20contains%201.689%20litres

LPG : https://www.elgas.com.au/elgas-knowledge-hub/residential-lpg/lpg-gas-unit-conversions/#:~:text=To%20convert%20kg%20to%20litres,LPG%20in%20a%2018kg%20cylinder.

Also, we have to do the conversion for the Electric Vehicle with the unit of km/Full Charge.

```python
df['Mileage_unit'] = df['Mileage_unit'].str.strip()
df[df["Mileage_unit"] == "km/Full Charge"]
```

When we do the internet research, we can find out that the energy consumption of Mahindra E Verito D4 is 13kwh/100km. 

Reference: https://www.pluginindia.com/mahindra-everito-review.html#:~:text=Efficiency%20&%20Range%20:%203%20out%20of%206&text=%5B+1%5D%20Calculated%20efficiency%20of,electric%20car%20for%20it's%20size

Also, since 1 gallon of gasoline is equal to 33.7 kwh of electric energy, we need to convert this to liter. by divide 33.7 to 3.785 (1 gallon to liter), which is about 8.902 kwh. Thus, we can convert km/full charge to kmpl by the conversion rule for mpge.

Reference: https://www.xcelerateauto.com/blog/mpge-explanation

```python
cng_density = 1.689  
lpg_density = 1.96  
gasoline_eq_kwh = 33.7  
liters_per_gallon = 3.78541 
battery_capacity_kwh = 18.55

def mpge_to_kmpl(battery_capacity_kwh, driving_range_km):
    range_miles = driving_range_km * 0.621371 
    
    # Calculate energy consumption in kWh per 100 miles
    energy_consumption_per_100_miles = (battery_capacity_kwh / range_miles) * 100
    
    # Convert energy consumption to kWh per 100 km
    energy_consumption_per_100_km = energy_consumption_per_100_miles / 0.621371
    
    # Convert kWh per 100 km to liters per 100 km
    liters_per_100_km = (energy_consumption_per_100_km / gasoline_eq_kwh) * liters_per_gallon
    
    # Convert liters per 100 km to kilometers per liter
    km_per_liter = 100 / liters_per_100_km
    
    return km_per_liter

def convert_kmpl(df):
    if df['Fuel_Type'] == 'CNG':
        return df['Mileage_Num'] * cng_density  # Convert km/kg to kmpl for CNG
    elif df['Fuel_Type'] == 'LPG':
        return df['Mileage_Num'] * lpg_density  # Convert km/kg to kmpl for LPG
    elif df['Fuel_Type'] == 'Electric' and df['Mileage_unit'] == 'km/Full Charge':
        driving_range_km = df['Mileage_Num']
        return mpge_to_kmpl(battery_capacity_kwh, driving_range_km)
    else:
        return df['Mileage_Num']  # No conversion for other fuel types (like Petrol)


df['Mileage_Num_Converted'] = df.apply(convert_kmpl, axis=1)
```

### 3.2 Detecting Outlier

Since none of them are normally distributed, we need to identy outliers by using IQR.

```python
import pandas as pd

# Numerical Columns 
columns = ['Car_Age', 'Kilometers_Driven', 'Mileage', 'Engine', 'Power', 'Seats', 'Price']

# We need to define function for calculating IQR step by step
def find_outliers(df, column):
    
    # step 1 : calculating 24% quantile and 75% quantile
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    
    # step 2 : Calculate IQR by subracting Q1 from Q3
    IQR = Q3 - Q1
    
    # Step 3 : Get the lower limit and upper limet
    LowerLimit = Q1 - 1.5 * IQR
    UpperLimit = Q3 + 1.5 * IQR
    
    # step 5: if the value is smaller or larger than the limits, we will add it on the outlier
    outlier = df[(df[column] < LowerLimit) | (df[column] > UpperLimit)]
    return outlier

    # step 6 : Iterate over each columns and get the outliers
for c in columns:
    outliers = find_outliers(df,c)
    
    # Step 7 : Count how many outliers in the column by using count function
    print(f'Columna Name: {c} \n Number of Outliers: {outliers.count()[0]}\n')
```

## 4. Pearson correlation coefficient and scatter plot

### Calculate pearson correlation from scratch

Pearson Correleation Coefficient formula

$$
r = \frac{\sum{(X_i - \bar{X})(Y_i - \bar{Y})}}{\sqrt{\sum{(X_i - \bar{X})^2} \sum{(Y_i - \bar{Y})^2}}}
$$


```python
# Exclude outliers and missing values

df_copy = df[columns].copy().dropna()


def remove_outliers(dataframe, column):

    Q1 = dataframe[column].quantile(0.25)
    Q3 = dataframe[column].quantile(0.75)
    
    IQR = Q3 - Q1

    LowerLimit = Q1 - 1.5 * IQR
    UpperLimit = Q3 + 1.5 * IQR

    non_outlier = dataframe[(dataframe[column] >= LowerLimit) & (dataframe[column] <= UpperLimit)]
    return non_outlier

for c in columns:
    filtered_df = remove_outliers(df_copy, c)

    
filtered_df.info()
```

```python
import itertools
import math
import pandas as pd

def Pcorr(data):
    corr_matrix = pd.DataFrame(index=data.columns, columns=data.columns)

    for c1, c2 in itertools.combinations(data.columns, 2):
        sumXY = 0
        Xsqrd = 0
        Ysqrd = 0
        
        for x, y in zip(data[c1], data[c2]):
            sumXY += (x - data[c1].mean()) * (y - data[c2].mean())
            Xsqrd += (x - data[c1].mean())**2
            Ysqrd += (y - data[c2].mean())**2
            
        sqrtXY = math.sqrt(Xsqrd * Ysqrd)
        
        if sqrtXY == 0:
            pearson_corr = 0  
        else:
            pearson_corr = sumXY / sqrtXY
        
        corr_matrix.at[c1, c2] = pearson_corr
        corr_matrix.at[c2, c1] = pearson_corr
    
    for col in data.columns:
        corr_matrix.at[col, col] = 1.0
    
    return corr_matrix
```

Manually Computed Pearson Correlation from Scratch

```python
correlation_matrix = Pcorr(filtered_df)
correlation_matrix
```

Pandas Pearson Correlation function used

```python
filtered_df.corr(method='pearson')
```

Although manually computed pearson correlation is a bit slower to compute the whole correlation, we can observe that the result is same as that of using Pandas function.

### Histogram

```python
import matplotlib.pyplot as plt
import numpy as np

def plot_histogram_with_imputed(df, df_copy, column):
    
    original_values = df_copy[column].dropna()
    imputed_values = df[column][df_copy[column].isnull()]

    plt.figure(figsize=(10, 6))
    plt.hist(original_values, bins=30, alpha=0.5, label='Original', color='blue')

    plt.hist(imputed_values, bins=30, alpha=1, label='Imputed', color='red')

    plt.title(f'Distribution of {column} (Imputed Values Highlighted)')
    plt.xlabel(column)
    plt.ylabel('Frequency')
    plt.legend()
    plt.show()

for c in ['Engine', 'Power', 'Seats']:
    plot_histogram_with_imputed(df, df_copy, c)
```
