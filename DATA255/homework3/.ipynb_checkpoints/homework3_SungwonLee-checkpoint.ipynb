## Step 1

````python
import os
import shutil
from pathlib import Path
from PIL import Image
# from many characters in our dataset
# choose only 3 custom categories (Homer Simpson, Marge Simpson, and Bart Simpson)
# and choose only 100 images each

parents_dir = "./data/simpsons_dataset/"
custom_dir = "simpsons_dataset_custom"

Path(custom_dir).mkdir(parents=True, exist_ok=True)
characters = ['bart_simpson', 'homer_simpson', 'marge_simpson']

for c in characters:
    char_parents_dir = os.path.join(parents_dir, c)
    char_custom_dir = os.path.join(custom_dir, c)
    Path(char_custom_dir).mkdir(parents=True, exist_ok=True)
    
    images = [file for file in os.listdir(char_parents_dir) if file.endswith(('.jpg', '.jpeg', '.png'))]
    images_selected = images[:100]
    
    for img in images_selected:
        src_path = os.path.join(char_parents_dir, img)
        dst_path = os.path.join(char_custom_dir, img)
        shutil.copyfile(src_path, dst_path)
````

## Step 2

````python
from torchvision import datasets, transforms
from torch.utils.data import DataLoader, random_split

data_transform = transforms.Compose([
    transforms.Resize((384, 256)),
    transforms.RandomHorizontalFlip(),
    transforms.RandomRotation(10),
    transforms.ToTensor(), # We are not going to do CenterCrop since some images' main target is not in the center
    transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])
])
````

````python
from torchvision.datasets import ImageFolder

dataset = ImageFolder(custom_dir, transform=data_transform)
train_size = int(0.8 * len(dataset))
val_size = len(dataset) - train_size
train_dataset, val_dataset = random_split(dataset, [train_size, val_size])
````

````python
print(f"Train dataset size: {len(train_dataset)}")
print(f"Test dataset size: {len(val_dataset)}")
````

## Step 4

````python
import torch
import torch.nn as nn

class CustomCNN(nn.Module):
    def __init__(self, num_classes):
        super(CustomCNN, self).__init__()
        self.rescale = nn.Sequential(
            nn.Conv2d(3, 3, kernel_size=1),
            nn.BatchNorm2d(3),
            nn.ReLU(inplace=True)
        )
        self.conv1 = nn.Conv2d(3, 16, kernel_size=3, padding=1)
        self.pool = nn.MaxPool2d(kernel_size=2, stride=2)
        self.conv2 = nn.Conv2d(16, 32, kernel_size=3, padding=1)
        self.conv3 = nn.Conv2d(32, 64, kernel_size=3, padding=1)
        self.fc1 = nn.Linear(64 * (384 // 8) * (256 // 8), 256)
        self.fc2 = nn.Linear(256, 128)
        self.fc3 = nn.Linear(128, num_classes)
        self.dropout = nn.Dropout(p=0.5)
        
    def forward(self, x):
        x = self.pool(F.relu(self.conv1(x)))
        x = self.pool(F.relu(self.conv2(x)))
        x = self.pool(F.relu(self.conv3(x)))
        x = x.view(-1, 64 * (384 // 8) * (256 // 8))
        x = F.relu(self.fc1(x))
        x = self.dropout(x)  # Apply dropout
        x = F.relu(self.fc2(x))
        x = self.fc3(x)
        return x
    
num_classes = 3
model = CustomCNN(num_classes)
print(model)
````

## Step 5

````python
import torch.optim as optim
import torch.nn.functional as F

optimizer = optim.Adam(model.parameters(), lr=0.001)
criterion = nn.CrossEntropyLoss()

def accuracy(outputs, labels):
    _, preds = torch.max(outputs, dim=1)
    return torch.tensor(torch.sum(preds == labels).item() / len(preds))

# Define the training loop
def train_model(model, train_dl, criterion, optimizer, num_epochs=10):
    for epoch in range(num_epochs):
        model.train()  # Set the model to train mode
        train_loss = 0.0
        train_acc = 0.0

        for images, labels in train_dl:
            optimizer.zero_grad()  # Zero the parameter gradients
            outputs = model(images)  # Forward pass
            loss = criterion(outputs, labels)  # Calculate the loss
            loss.backward()  # Backward pass
            optimizer.step()  # Optimize

            # Compute training accuracy
            acc = accuracy(outputs, labels)
            train_acc += acc.item() * images.size(0)

            # Track training loss
            train_loss += loss.item() * images.size(0)

        # Print training statistics
        epoch_loss = train_loss / len(train_dl.dataset)
        epoch_acc = train_acc / len(train_dl.dataset)
        print(f"Epoch {epoch + 1}/{num_epochs}, Loss: {epoch_loss:.4f}, Accuracy: {epoch_acc:.4f}")


num_epochs = 10
train_model(model, train_dataloader, criterion, optimizer, num_epochs)
````

Now we will test on the Test dataset

````python
correct = 0
total = 0
model.eval()

with torch.no_grad():
    for images, labels in val_dataloader:
        outputs = model(images)
        
        _, predicted = torch.max(outputs, 1)
        total += labels.size(0)
        correct += (predicted == labels).sum().item()
        
accuracy = correct / total * 100
print(f"Test Accuracy: {accuracy:.4f}&")
````

## Step 6

````python
import torchvision.models as models

googlenet = models.googlenet(pretrained=True)

for param in googlenet.parameters():
    param.requires_grad = False

googlenet.fc = nn.Linear(googlenet.fc.in_features, 3)

device = torch.device('cuda:0' if torch.cuda.is_available() else 'cpu')
googlenet = googlenet.to(device)

criterion = nn.CrossEntropyLoss()
optimizer = torch.optim.Adam(googlenet.fc.parameters(), lr=0.001)
````

## Step 7

````python
epochs = 10 

for epoch in range(epochs):
    googlenet.train()
    running_loss = 0.0
    
    for images, labels in train_dataloader:
        images, labels = images.to(device), labels.to(device) 
        optimizer.zero_grad()
        
        # Forward pass
        outputs = googlenet(images)
        loss = criterion(outputs, labels)
        
        # Backward pass and optimization
        loss.backward()
        optimizer.step()
        
        running_loss += loss.item()

    print(f'Epoch [{epoch+1}/{epochs}], Loss: {running_loss/len(train_dataloader):.4f}')
````

````python
correct = 0
total = 0
googlenet.eval()

with torch.no_grad():
    for images, labels in val_dataloader:
        images, labels = images.to(device), labels.to(device)
        outputs = googlenet(images)
        _, predicted = torch.max(outputs, 1)
        total += labels.size(0)
        correct += (predicted == labels).sum().item()

googlenet_accuracy = correct / total * 100
print(f'GoogleNet Validation Accuracy: {googlenet_accuracy:.2f}%')

print(f'Custom CNN Accuracy: {accuracy:.2f}%')
````
